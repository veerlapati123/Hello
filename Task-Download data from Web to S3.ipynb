```python
import json
import urllib.request
import boto
import boto.s3
import sys
from boto.s3.key import Key

#function to get the urls from json
def geturls(url):
    response = urllib.request.urlopen(url)
    data = response.read()      # a `bytes` object
    text = data.decode('utf-8') # a `str`; this step can't be used if data is binary
    json1_data = json.loads(text) #converting the string to json format
    json_dataset = json1_data['dataset'] #converting the json to dictionary

    url_csv = []
    #print all the download urls
    for x in json_dataset:
        try:
            temp_url = x['distribution'][0]['accessURL']
            #print(title)
            #print(type(temp_url))
            if "csv"not in str(temp_url):
                pass
            else:
                url_csv.append(temp_url)  
        except:
            pass
    return(url_csv)
#function to get the file upload status
def percent_cb(complete, total):
        sys.stdout.write('.')
        sys.stdout.flush()
    
    
#function to push files to s3
def pushfile2s3(AWS_ACCESS_KEY_ID,AWS_SECRET_ACCESS_KEY,bucket_name,testfile):
        print("Accesing the AWS....")

        conn = boto.connect_s3(AWS_ACCESS_KEY_ID,AWS_SECRET_ACCESS_KEY)
        print(bucket_name,"....")

        bucket = conn.create_bucket(bucket_name,location=boto.s3.connection.Location.DEFAULT)
        print('Uploading %s to Amazon S3 bucket %s' % (testfile, bucket_name))
        
        k = Key(bucket)
        k.key = testfile.split("/")[-1]
        k.set_contents_from_filename(testfile,cb=percent_cb, num_cb=10)
```
